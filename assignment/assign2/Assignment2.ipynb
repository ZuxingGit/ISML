## Q1

**Primal Form:**

The primal form of the hard-margin SVM is given by:

 $$\min_{\mathbf{w},b} \frac{1}{2} \|\mathbf{w}\|^2,\quad &\text{s.t.}\quad & y_{i}(\mathbf{w}^{\top} \mathbf{x}_{i} + b) \geq 1, \quad i=1,2,...,N\\$$



**Lagrangian:**

We introduce Lagrange multipliers **Î±** to formulate the Lagrangian for this problem. The Lagrangian is given by:

$$\mathcal{L} (\mathbf{w}, b, \alpha) = \frac{1}{2} \|\mathbf{w}\|^2 - \sum_{i=1}^{N}\alpha_{i}[y_{i}(\mathbf{w}^{\top} \mathbf{x}_{i} + b) - 1]$$

where $\alpha_{i} \geq 0, i=1,...,N$ are the Lagrange multipliers.



The objective is now:

$$\min_{\mathbf{w}, b}  \max_{\alpha_{i} \geq 0} \mathcal{L}(\mathbf{w}, b, \alpha) = \min_{\mathbf{w}, b} \max_{\alpha_{i} \geq 0} \left( \frac{1}{2} \|\mathbf{w}\|^2 - \sum_{i=1}^{n} \alpha_i [y_i (\mathbf{w}^T \mathbf{x}_i + b) - 1] \right)$$



**Simplify $$\mathcal{L} (\mathbf{w}, b, \alpha)$$:**

$$\mathcal{L} (\mathbf{w}, b, \alpha) = \frac{1}{2} \|\mathbf{w}\|^2 - \sum_{i=1}^{N}\alpha_{i}[y_{i}(\mathbf{w}^{\top} \mathbf{x}_{i} + b) - 1]$$

$$\quad\quad\quad\quad\quad = \frac{1}{2} \mathbf{w}^{\top}\mathbf{w} - \sum_{i=1}^{N}\alpha_{i}y_{i}\mathbf{w}^{\top} \mathbf{x}_{i} - \sum_{i=1}^{N}\alpha_{i}y_{i}b + \sum_{i=1}^{N}\alpha_{i}$$

$$\quad\quad\quad\quad\quad = \frac{1}{2} \mathbf{w}^{\top}\mathbf{w} - \mathbf{w}^{\top}\sum_{i=1}^{N}\alpha_{i}y_{i} \mathbf{x}_{i} - \sum_{i=1}^{N}\alpha_{i}y_{i}b + \sum_{i=1}^{N}\alpha_{i}$$



To find the optimal values for $$\mathbf{w}$$ and b, we differentiate the Lagrangian with respect to them and set the derivatives to zero:

$$\frac{\partial\mathcal{L}}{\partial \mathbf{w}} = \mathbf{w} - \sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i = 0 \implies \mathbf{w} = \sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i$$

$$\frac{\partial\mathcal{L}}{\partial b} = -\sum_{i=1}^{N} \alpha_i y_i = 0 \implies \sum_{i=1}^{N} \alpha_i y_i = 0$$



We have $$\sum_{i=1}^{N} \alpha_i y_i = 0$$, so $$\sum_{i=1}^{N}\alpha_{i}y_{i}b = 0$$. We can eliminate the term and get:

$$\mathcal{L} (\mathbf{w}, \alpha) = \frac{1}{2} \mathbf{w}^{\top}\mathbf{w} - \mathbf{w}^{\top}\sum_{i=1}^{N}\alpha_{i}y_{i} \mathbf{x}_{i} + \sum_{i=1}^{N}\alpha_{i}$$



We can substitute the expressions $$\mathbf{w} = \sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i$$  back into the Lagrangian and simplify further:

$$\mathcal{L} (\mathbf{w}, \alpha) = \frac{1}{2} \mathbf{w}^{\top}\sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i - \mathbf{w}^{\top}\sum_{i=1}^{N}\alpha_{i}y_{i} \mathbf{x}_{i} + \sum_{i=1}^{N}\alpha_{i}$$

$$\mathcal{L} (\alpha) = \sum_{i=1}^{N}\alpha_{i} - \frac{1}{2} \mathbf{w}^{\top}\sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i$$

$$\quad\quad\quad\quad\quad = \sum_{i=1}^{N}\alpha_{i} - \frac{1}{2} (\sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i)^{\top}\sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i$$

$$\quad\quad\quad\quad\quad = \sum_{i=1}^{N}\alpha_{i} - \frac{1}{2} (\sum_{i=1}^{N} \alpha_i y_i \mathbf{x}_i)^{\top}\sum_{j=1}^{N} \alpha_j y_j \mathbf{x}_j$$

$$\quad\quad\quad\quad\quad = \sum_{i=1}^{N}\alpha_{i} - \frac{1}{2} \sum_{i=1}^{N} \sum_{j=1}^{N}\alpha_i \alpha_j y_i y_j \mathbf{x}_i^{\top} \mathbf{x}_j$$



After simplification, the Lagrangian function only depends on $$\alpha$$, and $$\mathbf{w}$$ and b are eliminated from the formulation. Finally, we got $$\mathcal{L} (\alpha)$$ by rewriting the expression in terms of inner products:

$$\mathcal{L} (\alpha) = \sum_{i=1}^{N} \alpha_i - \frac{1}{2} \sum_{i=1}^{N} \sum_{j=1}^{N} \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j)$$



**Dual Formation:**
Now, we can write the dual form by maximising the Lagrangian with respect to $$\alpha$$ while subject to the constraints:

$$ \max_{\alpha}  \sum_{i=1}^{N}\alpha_{i} - \frac{1}{2}\sum_{i=1}^{N}\sum_{j=1}^{N}\alpha_{i}\alpha_{j}y_{i}y_{j}(\mathbf{x}_i \cdot \mathbf{x}_j)\\$$

$$\text{s.t.} \quad
 \alpha_{i} \geq 0, i=1,2,...,N$$

$$\quad \sum_{i=1}^{N}\alpha_{i}y_{i} = 0\\$$

## Q2

### Data Preprocessing

```python
import cvxpy as cp
import numpy as np
import pandas as pd

# Load data
train_df = pd.read_csv('train.csv', header = None)   
test_df = pd.read_csv('test.csv', header = None)

# Split training data 
X_train = train_df.iloc[:4000,1:].values
y_train = train_df.iloc[:4000,0].values
X_valid = train_df.iloc[4000:,1:].values
y_valid = train_df.iloc[4000:,0].values

# Test data
X_test = test_df.iloc[:1500,1:].values
y_test = test_df.iloc[:1500,0].values 

# change the label set from [0,1] to [-1,1]
y_train[y_train==0] = -1
y_valid[y_valid==0] = -1
y_test[y_test==0] = -1
```

### Solving

```python
def svm_train_primal(data_train, label_train, regularisation_para_C):
    N,M = np.shape(X_train)
    w = cp.Variable(M)
    b = cp.Variable()
    psi = cp.Variable(N)
    
    obj = cp.Minimize(1/2 * cp.square(cp.norm(w)) + regularisation_para_C/N * cp.sum(psi))
    constraints = [cp.multiply(label_train, cp.matmul(data_train, w) + b) - 1 + psi >= 0, psi >= 0]
    
    prob = cp.Problem(obj, constraints)
    prob.solve(solver = cp.ECOS)
    
    svm_model = [w.value, b.value]
    
    return svm_model

def svm_predict_primal(data_test, label_test, svm_model):
    pred = data_test @ svm_model[0] + svm_model[1]
    test_accuracy = (pred * label_test >= 0).mean()
    return test_accuracy
```

```python
svm_model = svm_train_primal(X_train, y_train, 100)

print('Sum of Weights:', np.sum(svm_model[0]))  
print('Bias:', svm_model[1])
print('Weights:', svm_model[0])

# Evaluate on validation set
print('Validation accuracy:', svm_predict_primal(X_valid, y_valid, svm_model))

# Evaluate on test set
print('Test accuracy:', svm_predict_primal(X_test, y_test, svm_model))
```

## Q3

```python
# Using CVXPY
def svm_train_dual(data_train, label_train, regularisation_para_C):
    N,M = np.shape(X_train)
    alpha = cp.Variable(N)
    
    obj = cp.Maximize(cp.sum(alpha) - (1/2) * cp.sum_squares(data_train.T @ (cp.multiply(alpha, label_train))))
    constraints = [alpha >= 0, alpha <= regularisation_para_C / N, cp.sum(cp.multiply(alpha, label_train)) == 0]
    
    prob = cp.Problem(obj, constraints)
    prob.solve(solver = cp.ECOS)

    return alpha.value

alpha = svm_train_dual(X_train, y_train, 100)

print('Sum of All Dimensions of Alpha:', np.sum(alpha))  
print('Alpha:', alpha)
```

## Q4

```python
import numpy as np

def obtain_w_b(data_train, label_train, alpha, regularisation_para_C):
    N, M = data_train.shape
    
    # calculate w
    w = np.zeros(M)
    for i in range(N):
        w += alpha[i] * label_train[i] * data_train[i]
        
    # calculate b from a support vector
    support_index = np.where((alpha >= 1e-6) & (regularisation_para_C / N - alpha >= 1e-6))[0]
    first = support_index[0]
    b = label_train[first] - np.dot(w, data_train[first])
    
    return w, b

svm_model = obtain_w_b(X_train, y_train, alpha, 100)
print('Sum of w*:', np.sum(svm_model[0]))  
print('b*:', svm_model[1])
print('w*:', svm_model[0])

# Evaluate on validation set
print('Validation accuracy:', svm_predict_primal(X_valid, y_valid, svm_model))

# Evaluate on test set
print('Test accuracy:', svm_predict_primal(X_test, y_test, svm_model))
```

## Q5

```python
def find_support_vectors_primal(data_train, label_train, regularisation_para_C):
    support_vectors = []
    N, M = data_train.shape
    
    svm_model = svm_train_primal(data_train, label_train, regularisation_para_C)
    w, b = svm_model[0], svm_model[1]
    
    for i in range(N):
        if label_train[i] == 1:  
            if np.abs(np.dot(w, data_train[i]) + b - 1) <= 1e-6:
                support_vectors.append(data_train[i])
        if label_train[i] == -1:  
            if np.abs(np.dot(w, data_train[i]) + b + 1) <= 1e-6:
                support_vectors.append(data_train[i])
    return support_vectors

support_vectors_primal = find_support_vectors_primal(X_train, y_train, 100)
print("Support Vectors from the Primal Problem Solutions:")
support_vectors_primal
```

## Q3 Using CVXOPT

```python
import cvxopt
import cvxopt.solvers as solvers
import numpy as np

# Using CVXOPT
 
def svm_train_dual(data_train, label_train, regularisation_para_C):
    N, M = data_train.shape
    
    K = np.dot(data_train, data_train.T)
    
    P = cvxopt.matrix(np.outer(label_train, label_train) * K)
    q = cvxopt.matrix(-np.ones((N, 1)))
    G = cvxopt.matrix(np.vstack((np.eye(N) * (-1), np.eye(N))))
    h = cvxopt.matrix(np.hstack((np.zeros(N), np.ones(N) * (regularisation_para_C / N))))
    A = cvxopt.matrix(label_train, (1,N))
    b = cvxopt.matrix(np.zeros(1))
    
    sol = solvers.qp(P, q, G, h, A, b)
    
    alpha = np.ravel(sol['x'])
    return alpha

alpha = svm_train_dual(X_train, y_train, 100)

print('Sum of All Dimensions of Alpha:', np.sum(alpha))  
print('Alpha:', alpha)
```
